# Scenario:
 You are a data science trainer preparing a demo for a group of first-time managers in e-commerce.

 You want to show them how to securely connect to Azure OpenAI from Google Colab, send a simple query,
  and interpret the response. The demo emphasizes security (using secrets) and clarity (structured output).


```python
from openai import OpenAI

api_key = userdata.get("OPENAI_API_KEY")
endpoint = "https://vivek-18-resource.openai.azure.com/openai/v1/"
deployment_name = "gpt-oss-120b"

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "What is the capital of France?",
        }
    ],
)

print(completion.choices[0].message)
```

# Scenario
You’re training healthcare analysts.

You want them to see how to connect securely to Azure OpenAI, ask questions about AI in healthcare, and then deploy the interaction into a Gradio interface so they can try it themselves.

```python
from openai import OpenAI
import gradio as gr
import os

api_key = userdata.get("OPENAI_API_KEY")

endpoint = "https://vivek-18-resource.openai.azure.com/openai/v1/"
deployment_name = "gpt-oss-120b"

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

def healthcare_chat(user_input):
    try:
        response = client.chat.completions.create(
            model=deployment_name,
            messages=[
                {
                    "role": "system",
                    "content": "You are a healthcare AI assistant helping analysts understand AI applications in healthcare."
                },
                {
                    "role": "user",
                    "content": user_input
                }
            ],
            temperature=0.7
        )

        return response.choices[0].message.content

    except Exception as e:
        return f"Error: {str(e)}"


interface = gr.Interface(
    fn=healthcare_chat,
    inputs=gr.Textbox(
        lines=3,
        placeholder="Ask about AI in healthcare (e.g., diagnosis, patient data, NLP in medical records...)"
    ),
    outputs=gr.Textbox(lines=12),
    title="Healthcare AI Assistant",
    description="Ask questions about AI applications in healthcare."
)

if __name__ == "__main__":
    interface.launch()
```
